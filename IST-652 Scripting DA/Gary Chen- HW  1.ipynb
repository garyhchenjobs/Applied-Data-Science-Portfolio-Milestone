```python
import pandas as pd

ramen_file = 'ramen-ratings.csv' #path for ramen-ratings.csv

#Read Ramen file into dataframe, csv = comma as the separator 

ramen_df = pd.read_csv(ramen_file, sep=',')
ramen_df.shape
```

```python
#top 5 rows of ramen df
ramen_df.head()
```

```python
#get descriptive statistics from numeric columns

reviews = ramen_df['Review #']
stars = ramen_df['Stars']

print(reviews.describe())

print('\n')

print(stars.describe())

#not getting full descriptive statistics on Stars column - dtype: object
```

```python
#check distinct values for stars column
unique_star = stars.unique()
print(unique_star)
```

```python
#Found 'Unrated' which is not a int or float
#convert 'Unrated' to 0.001 to designate as Unrated
#This will allow us to do calculations with the 'Stars' column
ramen_df.loc[ramen_df['Stars'] == 'Unrated', "Stars"] = 0.001

print(ramen_df['Stars'])
print('\n')

#convert Stars column to numeric to do calculations
ramen_df['Stars'] = pd.to_numeric(ramen_df['Stars'])

#summary stats for 'Stars' column
print(ramen_df['Stars'].describe())
```

```python
#Which country has the most 5.0 stars rated ramens?

#first create a new ramen df with only 5.0 stars
Ramen_5stars = ramen_df[ramen_df['Stars'] == 5.000]
display(Ramen_5stars)

print('\n')

#do the value counts number of 5 stars by country
print(Ramen_5stars['Country'].value_counts())
```

```python
#check for NULL values for Top Ten 

ramen_df[ramen_df['Top Ten'].isnull()]

ramen_df['Top Ten'].isnull()
```

```python
#update Top Ten NUlL values to no 
ramen_df.loc[ramen_df['Top Ten'].isnull(), 'Top Ten'] = 'no'

ramen_df.head()
```

```python
#create 'Ramen Score' column based on Review #, Stars, & Top Ten flag

#first create a 'Top Ten points score column - If Ramen was rated top ten of a year, 500 points. If not, 0 points.
#This will give a yearly top 10 rated Ramen extra 500 pts to be calculated in our 'Ramen Score'
#500 extra pts was caculated by a Top Ten award should equal an additional 100 reviews of 5 stars
#We will not scale down these 100 reviews like we will do later
#This will add 500 pts to a yearly Top Ten awarded Ramen's Score

#import numpy to use where function
import numpy as np

#check all Top Ten values that do not equal to no
display(ramen_df[ramen_df['Top Ten'] != 'no'])

print('\n')

#there are some 'Top Ten' values as \n. We can replace that with no. I don't think '\n' denotes as Top Ten ramen

#update Top Ten '\n' values with no
ramen_df.loc[ramen_df['Top Ten'] == '\n', 'Top Ten'] = 'no'
#display & check Top Ten values ago 
display(ramen_df[ramen_df['Top Ten'] != 'no'])

#create df fir awarded ramen
awarded_ramen = ramen_df[ramen_df['Top Ten'] != 'no']

#caclulate countries with top 
display(awarded_ramen['Country'].value_counts())

print('\n')

#use numpy where function to create 'Top Ten - extra pts' columns based on the condition
ramen_df['Top Ten - extra pts'] = np.where(ramen_df['Top Ten'] != 'no', 500, 0)

#check if Top Ten rated has been awarded the extra pts
display(ramen_df[ramen_df['Top Ten'] != 'no'])
```

```python
#view if Top Ten Ramen that have been appointed extra pts

ramen_df[ramen_df['Top Ten'] != 'no']
```

```python
#calculate Ramen Score by  ('Review #'/(3.65 * 1.5)) * 'Stars' + 'Top Ten - extra pts'
#dividing the Review number by 3.65 will scale down the weight of number of reviews that goes into calculating the ramen score.
#3.65 was choosen because that is mean/avg stars in the data set
#After using 3.65 as the divider to scale down, It still rewarded and weight Review # too much or too highly
#We are going to divide Review # by 5.475 instead. We got 5.475 from multiplying 3.65 by 1.5

ramen_df['Ramen Score'] = ((ramen_df['Review #'] / (3.65*1.5)) *  ramen_df['Stars'] + ramen_df['Top Ten - extra pts'])

display(ramen_df.head())

print('\n')

#get summary statistics of Ramen Score
display(ramen_df['Ramen Score'].describe())

print('\n')

#check the scores of Top Ten awarded Ramen
display(ramen_df[ramen_df['Top Ten'] != 'no'])

#This shows most of the top 10 awarded Ramen are above the mean Ramen Score.
#While the scoring system isn't perfect, it is a good way to  
```

```python
#Overall Mean Ramen Score by Country
Avg_Ramen_by_country = ramen_df.groupby(['Country']).mean('Ramen Score').sort_values(by = 'Ramen Score', ascending = False, inplace= False)
display(Avg_Ramen_by_country)
```
