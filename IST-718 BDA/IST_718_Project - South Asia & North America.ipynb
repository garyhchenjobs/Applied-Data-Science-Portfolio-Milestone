```python
# restrict livestock to North America
live_north = live.loc[live['Area'] == 'Northern America']
live_north.head()
```

```python
live_north.loc[live_north['Item'] == 'Cattle'].head()
```

```python
live_north['Area'] = live_north['Area'].replace(['Northern America'], 'North America')
```

```python
live_north.rename(columns={'Area': 'Region', 'Item': 'Animal species'}, inplace=True)
live_north.head()
```

```python
# join emissions and livestock dataframes
live_emissions = pd.merge(live_north, emissions, on=['Region', 'Animal species'], how = 'inner')
live_emissions.loc[live_emissions['Animal species'] == 'Cattle'].head()
```

```python
# restrict livestock to South Asia
live_south_asia = live.loc[live['Area'] == 'Southern Asia']
live_south_asia.head()
```

```python
live_south_asia['Area'] = live_south_asia['Area'].replace(['Southern Asia'], 'South Asia')
live_south_asia['Item'] = live_south_asia['Item'].replace(['Chickens'], 'Chicken')
```

```python
live_emissions_south_asia = pd.merge(live_south_asia, emissions_south_asia
                                     , left_on=['Area', 'Item'], right_on = ['Region', 'Animal species'], how='inner')

live_emissions_south_asia.head()
#live_emissions_south_asia.loc[live_emissions['Animal species'] == 'Cattle'].head()
```

```python
# kg protein per head x value
# cattle 52 kg per head
# chicken 0.165kg per chicken
# sheep 20 kg protein per head
# buffalo 112 kg per head
# goats 5-7 kg protein per head
# pigs 12 kg protein per head

# join emissions and livestock dataframes
live_emissions_south_asia["Value"] = np.where(live_emissions_south_asia["Animal species"] == "Chicken"
                                              , live_emissions_south_asia["Value"]*1000*.165
                                              , live_emissions_south_asia["Value"])
live_emissions_south_asia["Value"] = np.where(live_emissions_south_asia["Animal species"] == "Cattle"
                                              , live_emissions_south_asia["Value"]*52
                                              , live_emissions_south_asia["Value"])
live_emissions_south_asia["Value"] = np.where(live_emissions_south_asia["Animal species"] == "Pigs"
                                              , live_emissions_south_asia["Value"]*12
                                              , live_emissions_south_asia["Value"])
live_emissions_south_asia["Value"] = np.where(live_emissions_south_asia["Animal species"] == "Buffaloes"
                                              , live_emissions_south_asia["Value"]*112
                                              , live_emissions_south_asia["Value"])
live_emissions_south_asia["Value"] = np.where(live_emissions_south_asia["Animal species"] == "Sheep"
                                              , live_emissions_south_asia["Value"]*20
                                              , live_emissions_south_asia["Value"])
live_emissions_south_asia["Value"] = np.where(live_emissions_south_asia["Animal species"] == "Goats"
                                              , live_emissions_south_asia["Value"]*6
                                              , live_emissions_south_asia["Value"])
live_emissions_south_asia['Final_GHG'] = live_emissions_south_asia['Value'] * live_emissions_south_asia['Emission Intensity (kg CO2e per kg protein)']
# live_emissions.loc[live_es_asia_sum['Animal species'] == 'Cattle'].head()
live_emissions_south_asia['Final_GHG'].value_counts
```

```python
#prophet DF clean for South Asia livestock emissions
live_prophet_SA = live_emissions_south_asia[['Animal species', 'Year', 'Final_GHG']]
live_prophet_SA = live_prophet_SA.rename(columns ={'Animal species':'group', 'Year':'ds', 'Final_GHG':'y'})
live_prophet_SA['ds'] = pd.to_datetime(live_prophet_SA['ds'], format='%Y')

display(live_prophet_SA)
```

```python
from prophet import Prophet
# Create a list of unique/distinct livestock
groups = live_emissions_south_asia['Animal species'].unique()

# Create a separate forecast for each group with iterating over unique zipcode groups
forecasts = []

for group in groups:
    
    # Subset the data for each group
    group_data = live_prophet_SA[live_prophet_SA['group'] == group]

    # Create Prophet model
    model = Prophet()
    model.fit(group_data)

    # Precdict forecast for the next 10 Years
    future = model.make_future_dataframe(periods=10, freq='Y')
    forecast = model.predict(future)

    # Add the group value column to the forecast DataFrame
    forecast['group'] = group

    # Append the forecast to the list of forecasts
    forecasts.append(forecast)
    
    #print unaltered axis of each forecast as part of the for-loop
    model.plot(forecast)
    plt.legend([group,'forecast', 'uncertainty'],loc='upper left')
    
    #print second copy of each forecast with normalized axis
    model.plot(forecast)
    plt.ylim([1e4, 4e12])
    plt.legend([group,'forecast', 'uncertainty'],loc='upper left')

# Combine the forecasts for all groups into a single DataFrame
combined_forecast = pd.concat(forecasts)

# display(combined_forecast)
```
