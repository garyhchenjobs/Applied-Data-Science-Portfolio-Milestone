```python
import csv
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ghg_df = pd.read_csv('ghg-transposed.csv')
ghg_df.head()
```

```python
ghg_sort = ghg_df.sort_values(by=['1990'], ascending=False)
ghg_top12 = ghg_sort.head(12)
ghg_top12
```

```python
ghg_box = ghg_top12.iloc[: , 1:]
ax = sns.boxplot(ghg_top12)
ax.set_title("GHG Variation / Year",color='black')
ax.tick_params(axis='x', rotation=90)
ax.plot()
```

```python
ghg_transposed = ghg_top12.set_index('Country/Region').transpose()
ghg_transposed.head()
```

```python
sns.heatmap(ghg_transposed, cmap = "Greens")
```

```python
ax = ghg_transposed.plot.line()
ax.set_title("GHG / Year / Country",color='black')
ax.legend(bbox_to_anchor=(1.0, 1.0))
ax.plot()
```

```python
global_df = pd.read_csv('GLEAM_global.csv')
global_df.head()
```

```python
sns.heatmap(global_df.corr())
```

```python
ax = sns.barplot(global_df, x="Emission Intensity (kg CO2e per kg protein)", y="Production (kg protein)",
                 hue="CONCAT", width=1)
ax.set_title("Production / Year / Country",color='black')
ax.legend(bbox_to_anchor=(1.0, 1.0))
ax.tick_params(axis='x', rotation=90)
ax.plot()
```
